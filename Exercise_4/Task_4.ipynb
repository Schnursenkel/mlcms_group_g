```python
# get p(x) - here by simple random generation
np.random.seed(100)# 100 is the seed used for the picture in the report, try 10 and 1 for drastic changes
x_bound = 190 # upper bound for the x coordinate
y_bound = 110 # upper bound for the y coordinate
num_samples = 1000
p_x_simple = np.random.rand(num_samples, 2) * np.array([x_bound, y_bound])
```

```python
np.shape(p_x_simple)# should be (num_samples, 2) i.e. 1000 samples of (x,y)-coordinates
```

```python
# plot the samples - will be totally random, no MI building at all
plt.scatter(p_x_simple[:,0],p_x_simple[:,1]);
plt.title('Random coordinate sampling with 1000 samples');
plt.xlabel('x');
plt.ylabel('y');
```

```python
# count the number of persons in the measurement area
counter = 0
for x,y in p_x_simple:
    if 130 <= x <= 150 and 50 <= y <= 70:
        counter += 1
print('Number of persons in the measurement area: '+str(counter))
```

```python
# calculate the expected number of samples that need to be drawn in order to have 100 persons in the measurement area
crit_num_persons = 100
crit_num_samples = math.ceil(crit_num_persons * num_samples / counter)
print('With the given distribution '+str(crit_num_samples)+' samples are needed to reach '+str(crit_num_persons)+' persons in the measurement area.')
```
