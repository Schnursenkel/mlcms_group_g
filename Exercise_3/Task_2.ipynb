```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from numpy import linalg as LA
import matplotlib.gridspec as gridspec
```

Exercise 3 task 2

```python
import cmath
#main equation x_dot=alpha-x**2
#node x_node=(+)(-)sqrt(alpha)
alpha = np.arange(-1.0, 1.0, 0.01)
alpha=np.array(alpha)
```

```python
list_n=[]
list_p=[]
for i in alpha:
    x_node_plus=cmath.sqrt(i)
    x_node_minus=-cmath.sqrt(i)
    list_n.append(x_node_minus)
    list_p.append(x_node_plus)
plt.plot(alpha,list_p,color='blue',label="when x0=sqrt(a)")
plt.plot(alpha,list_n,color='red',label="when x0=-sqrt(a)")
plt.legend()
plt.show()
```

```python
#main equation x_dot=alpha-2*x**2-3
#node x_node=(+)(-)sqrt((alpha-3)/2)
alpha = np.arange(-1.0, 4.0, 0.01)
alpha=np.array(alpha)
```

```python
list_n=[]
list_p=[]
for i in alpha:
    x_node_plus=cmath.sqrt((i-3)/2)
    x_node_minus=-cmath.sqrt((i-3)/2)
    list_n.append(x_node_minus)
    list_p.append(x_node_plus)
plt.plot(alpha,list_p,color='blue',label="when x0=sqrt((a-3)/2)")
plt.plot(alpha,list_n,color='red',label="when x0=-sqrt((a-3)/2)")
plt.legend()
plt.show()
```

Behaviour at alpha=0.0

```python
w = 3
alpha=0.0
Y, X = np.mgrid[-w:w:25j, -w:w:25j]
U = alpha-X**2
V = -Y
seed_points = np.array([[3,0,3,0,-1,-1,-1], [-3,-3,3,3,-3,0,3]])

fig = plt.figure(figsize=(15, 15))
gs = gridspec.GridSpec(nrows=3, ncols=2, height_ratios=[1, 1, 2])

#  Varying density along a streamline
ax0 = fig.add_subplot(gs[0, 0])
ax0.streamplot(X, Y, U, V, density=[0.5, 1])
ax0.streamplot(X, Y, U, V, color='orange', linewidth=2,
                         cmap='autumn', start_points=seed_points.T)
plt.plot(0,0,'rx') 
ax0.set_title('alpha=0')
```
