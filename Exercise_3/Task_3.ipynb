Task 3: Part 1: Visualize the bifurcation of the system by plotting three phase diagrams at representative values of alpha.


```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from numpy import linalg as LA
```

```python
import matplotlib.gridspec as gridspec
```

xdot_1=a*x1 - x2 - x1*(x1^2 + x2^2)

xdot_2=x1 + a*x2 - x2*(x1^2 + x2^2)

```python
def x_prime_1(alpha,x1,x2):
    x_prime_1=np.round(alpha*x1 - x2 - x1*(x1**2 + x2**2),10)
    return x_prime_1
```

```python
def x_prime_2(alpha,x1,x2):
    x_prime_2=np.round(x1 + alpha*x2 - x2*(x1**2 + x2**2),10)
    return x_prime_2
```

For alpha = -1.5

```python
alpha1=-1.5
alpha2=0.0
alpha3=1.5
w = 2
Y, X = np.mgrid[-w:w:25j, -w:w:25j]


fig = plt.figure(figsize=(10, 20))
gs = gridspec.GridSpec(nrows=3, ncols=2, height_ratios=[1, 1, 2])

#  Varying density along a streamline
seed_points = np.array([[1.0,0.5,0.0,0.0], [0.0,0.0,0.5,1.0]])
U1= x_prime_1(alpha1,X,Y)
V1=x_prime_2(alpha1,X,Y)
ax0 = fig.add_subplot(gs[1, 0])
ax0.quiver(X,Y,U1,V1)
ax0.streamplot(X, Y, U1,V1, start_points=seed_points.T,color='r')
ax0.set_title('Bifurcation at alpha {}, alpha < 0 case'.format(alpha1))
```

for alpha = 0.0

```python
w = 2
Y, X = np.mgrid[-w:w:25j, -w:w:25j]

seed_points = np.array([[1.0,0.5,0.0,0.0], [0.0,0.0,0.5,1.0]])
fig = plt.figure(figsize=(10, 20))
gs = gridspec.GridSpec(nrows=3, ncols=2, height_ratios=[1, 1, 2])

U2= x_prime_1(alpha2,X,Y)
V2=x_prime_2(alpha2,X,Y)
ax1 = fig.add_subplot(gs[1, 0])
ax1.quiver(X,Y,U2,V2)
ax1.streamplot(X, Y, U2,V2, start_points=seed_points.T,color='r')
ax1.set_title('Bifurcation at alpha {}, alpha = 0 case'.format(alpha2))
```

Task 3: Part 2 --> For alpha = 1, numerically compute and visualize two orbits of the system forward in time, starting at the
point (2; 0) and at (0.5; 0).

for starting point (2,0) plot is shown in red color

for starting point (0.5,0) plot is shown in blue color

```python
alpha=1.0
w = 2
Y, X = np.mgrid[-w:w:25j, -w:w:25j]
seed_points_1 = np.array([[2.0], [0.0]])
seed_points_2 = np.array([[0.5], [0.0]])
fig = plt.figure(figsize=(10, 20))
gs = gridspec.GridSpec(nrows=3, ncols=2, height_ratios=[1, 1, 2])

U= x_prime_1(alpha,X,Y)
V=x_prime_2(alpha,X,Y)

ax2 = fig.add_subplot(gs[1, 0])
ax2.quiver(X,Y,U,V)
ax2.streamplot(X, Y, U,V, start_points=seed_points_1.T,color='r')
ax2.streamplot(X, Y, U,V, start_points=seed_points_2.T,color='b')
ax2.set_title('Orbits at alpha {}, red for (2.0,0.0) blue for (0.5,0.0)'.format(alpha))
```

```python
x_range=np.arange(-1.0, 1.0, 0.1)
alpha_1=np.arange(-1.0,1.0, 0.1)
alpha_2=np.arange(-1.0, 1.0, 0.1)
print(x_range.shape)
print(alpha_1.shape)
print(alpha_2.shape)
```

```python
list_a1=[]
list_a2=[]
list_x=[]
for x in x_range:
    for a_1 in alpha_1:
        for a_2 in alpha_2:
            alpha1=x**3-a_2*x
            alpha2=(-a_1+x**3)/x
            list_x.append(x)
            list_a1.append(alpha1)
            list_a2.append(alpha2)
```

```python
fig,ax = plt.subplots(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(list_a1, list_a2, list_x,marker='o')
ax.set_xlabel('alpha_1')
ax.set_ylabel('alpha_2')
ax.set_zlabel('X')

plt.show()
```
